# State wise analysis of Profit,Discount,Sale

```python
Data['State'].value_counts().head(10)
```

```python
Data_state=Data.groupby(['State'])[['Sales','Discount','Profit']].mean()
Data_state.head(10)
```

```python
df_state1=Data_state.sort_values('Profit')
```

```python
df_state1[['Profit']].plot(kind='bar',figsize=(15,4))
plt.title('State wise Profit Analysis',fontsize=20)
plt.ylabel('Profit per State')
plt.xlabel('States')
plt.show()
```

Inference:
    * vermont : Highest Profit
    * Ohio : Lowest Profit

```python
Data_state['Sales'].plot(kind='pie',
                        figsize=(18,18),
                        autopct='%1.1f%%',
                        startangle=90,
                        shadow=True)
plt.title('State wise analysis of Sale',fontsize=20)
```

Inference: Highest amount of sales = Wyoming (11.8%)
           Lowest amount of sales = South Dakota(0.8%)

```python
df_state1['Discount'].plot(kind='bar',figsize=(18,5))
plt.title('State wise analysis of Discount',fontsize=20)
```

Inference: llinois is at top
